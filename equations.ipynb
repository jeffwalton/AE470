## Two-Body Orbits

\begin{equation}
\boxed{ \ddot{\vec{\mathbf{r}}} = - \frac{\mu}{|\vec{\mathbf{r}}|^3} \vec{\mathbf{r}} }
\end{equation}

\begin{equation}
\boxed{ \vec{\mathbf{h}} \triangleq \vec{\mathbf{r}} \times \vec{\mathbf{v}} }
\end{equation}

\begin{equation}
\boxed{ \mathcal{E} \triangleq \frac{v^2}{2} - \frac{\mu}{r} = \frac{\vec{\mathbf{v}} \cdot \vec{\mathbf{v}}}{2} - \frac{\mu}{r} 
 }
\end{equation}

\begin{equation}
\boxed{ \vec{\mathbf{e}} \triangleq \frac{ \dot{\vec{\mathbf{r}}} \times \vec{\mathbf{h}} }{\mu} - \frac{ \vec{\mathbf{r}}}{r} }
\end{equation}

\begin{equation}
 \boxed{r  = \frac{p}{1 + e \cos{\theta}}} 
\end{equation}

\begin{equation}
\boxed{ p = \frac{h^2}{\mu} }
\end{equation}

\begin{equation}
a = \bigg( \frac{2}{r} - \frac{ v^2 }{\mu} \bigg)^{-1}
\end{equation}

\begin{equation}
\boxed{ a = \frac{p}{1-e^2} }
\end{equation}

\begin{equation}
\boxed{ \mathcal{E} = -\frac{\mu}{2 a} }
\end{equation}

\begin{equation}
\boxed{ \mathcal{E} = \frac{\mu^2 (e^2 - 1)}{2 h^2} }
\end{equation}

\begin{equation}
\boxed{ \mathcal{E} = \frac{\mu^2 (e^2 - 1)}{2 h^2} }
\end{equation}

\begin{equation}
\boxed{ e = \sqrt{ 1+ \frac{2 \mathcal{E} h^2}{\mu^2} } }
\end{equation}

\begin{equation}
\boxed{ v = \sqrt{ \mu \bigg( \frac{2}{r} - \frac{1}{a} \bigg) } }
\end{equation}

### Circular Orbits

\begin{equation}
\boxed{ v_{circular} = \sqrt{ \frac{\mu}{r} } }
\end{equation}

\begin{equation}
\boxed{ T_{circular} = \frac{2 \pi r}{\sqrt{ \mu / r } } = \frac{2 \pi }{\sqrt{ \mu} }r^{\frac{3}{2}} }
\end{equation}

\begin{equation}
\boxed{ \mathcal{E}_{circular} = -\frac{\mu}{2 r} }
\end{equation}

\boxed{ r_p + r_a = 2 a }
\end{equation}

\end{equation}

\boxed{ T^2 = \frac{ 4 \pi^2 }{\mu}a^3 }
\end{equation}

\begin{equation}
